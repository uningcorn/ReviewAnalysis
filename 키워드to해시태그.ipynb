```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# 데이터 로드 (이미 데이터프레임 형태로 있는 것으로 가정)
df = pd.read_csv('test.csv') # 필요한 경우 파일에서 데이터를 불러올 수 있음

# 'keyword'와 'hashtag' 컬럼 사용
X = df['keyword'].fillna('')  # NaN 값을 빈 문자열로 대체
y = df['Hashtag'].fillna('')  # NaN 값을 빈 문자열로 대체

# TF-IDF 벡터화 (키워드를 숫자로 변환)
vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X)

# 데이터를 학습용/테스트용으로 분할 (X는 벡터화된 데이터, X_original은 원래 키워드) ->origin 삭제 
X_train, X_test, y_train, y_test, X_train_original, X_test_original = train_test_split(X_vec, y, X, test_size=0.2, random_state=42)

# SVM 분류 모델 생성 및 학습
model = SVC(kernel='linear')  # 선형 커널을 사용한 SVM
model.fit(X_train, y_train)

# 테스트 데이터로 예측
y_pred = model.predict(X_test)

# 정확도 계산
accuracy = accuracy_score(y_test, y_pred)
print(f"모델 정확도: {accuracy * 100:.2f}%")

# 예측 결과 확인 (원래 키워드와 예측된 해시태그 출력)
print("\n예측된 해시태그 예시:")
for i in range(5):  # 테스트 데이터의 상위 5개만 출력
    original_keyword = X_test_original.iloc[i]  # 원래 키워드 가져오기
    predicted_hashtag = y_pred[i]  # 예측된 해시태그 가져오기
    print(f"키워드: {original_keyword}, 예측된 해시태그: {predicted_hashtag}")
```

# LDA

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# 예시 리뷰 데이터
reviews = df['Review'].fillna('')

# 단어 빈도수 벡터화
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(reviews)

# LDA 모델 학습
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(X)

# 각 리뷰에 대한 토픽 분포 출력
topics = lda.transform(X)
print("각 리뷰의 토픽 분포:")
print(topics)
```

```python
# 주요 단어 출력 (각 토픽에서 중요한 단어)
for i, topic in enumerate(lda.components_):
    print(f"토픽 {i + 1}:")
    print([vectorizer.get_feature_names_out()[index] for index in topic.argsort()[-10:]])
```
