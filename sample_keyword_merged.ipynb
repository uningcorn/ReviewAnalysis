```python
df_final2['keyword_x'] = df_final2['keyword_x'].str.split(',')  # 첫 번째 키워드 열
df_final2['keyword_y'] = df_final2['keyword_y'].str.split(',')  # 두 번째 키워드 열
df_final2['keyword_df1'] = df_final2['keyword_df1'].str.split(',')  # 세 번째 키워드 열
df_final2['keyword_df2'] = df_final2['keyword_df2'].str.split(',')  # 세 번째 키워드 열

def find_common_keywords(row):
    # 각 열의 키워드를 set으로 변환
    set_x = set(row['keyword_x']) if isinstance(row['keyword_x'], list) else set()
    set_y = set(row['keyword_y']) if isinstance(row['keyword_y'], list) else set()
    set_df1 = set(row['keyword_df1']) if isinstance(row['keyword_df1'], list) else set()
    set_df2 = set(row['keyword_df2']) if isinstance(row['keyword_df2'], list) else set()
    
    # 네 열 모두에 공통으로 있는 키워드를 추출
    common_keywords = set_x.intersection(set_y).intersection(set_df1).intersection(set_df2)
    
    # 리스트를 다시 문자열로 변환해서 반환
    return ','.join(common_keywords)

# 공통된 키워드를 저장할 새 열 추가
df_final2['common_keywords'] = df_final2.apply(find_common_keywords, axis=1)

df_final2
```

```python
import pandas as pd

# 각 키워드 열을 리스트로 변환 (콤마로 구분된 키워드라고 가정)
df_final2['keyword_x'] = df_final2['keyword_x'].str.split(',')  # 첫 번째 키워드 열
df_final2['keyword_y'] = df_final2['keyword_y'].str.split(',')  # 두 번째 키워드 열
df_final2['keyword_df1'] = df_final2['keyword_df1'].str.split(',')  # 세 번째 키워드 열
df_final2['keyword_df2'] = df_final2['keyword_df2'].str.split(',')  # 네 번째 키워드 열

# 공통된 키워드를 찾는 함수 (4개 다 겹치는 것과 3개만 겹치는 것)
def find_keywords(row):
    # 각 열의 키워드를 set으로 변환
    set_x = set(row['keyword_x']) if isinstance(row['keyword_x'], list) else set()
    set_y = set(row['keyword_y']) if isinstance(row['keyword_y'], list) else set()
    set_df1 = set(row['keyword_df1']) if isinstance(row['keyword_df1'], list) else set()
    set_df2 = set(row['keyword_df2']) if isinstance(row['keyword_df2'], list) else set()
    
    # 모든 키워드 세트를 합침
    all_keywords = set_x.union(set_y).union(set_df1).union(set_df2)
    
    # 4개 열 모두에 공통으로 있는 키워드 추출
    common_all = set_x.intersection(set_y).intersection(set_df1).intersection(set_df2)
    
    # 3개 열에서만 공통으로 있는 키워드 추출
    common_3_out_of_4 = [keyword for keyword in all_keywords if (
        (keyword in set_x) + (keyword in set_y) + (keyword in set_df1) + (keyword in set_df2)
    ) == 3]
    
    # 결과를 문자열로 변환해서 반환
    return ','.join(common_all), ','.join(common_3_out_of_4)

# 공통된 키워드를 저장할 두 개의 새 열 추가
df_final2['common_keywords_all'] = df_final2.apply(lambda row: find_keywords(row)[0], axis=1)  # 4개 모두 겹치는 키워드
df_final2['common_keywords_3_out_of_4'] = df_final2.apply(lambda row: find_keywords(row)[1], axis=1)  # 3개 겹치는 키워드

# # 결과 확인
# print(df_final2[['keyword_x', 'keyword_y', 'keyword_df1', 'keyword_df2', 'common_keywords_all', 'common_keywords_3_out_of_4']].head())

# # 결과 저장
# df_final2.to_csv('df_with_common_keywords_all_and_3_out_of_4.csv', index=False)
```
