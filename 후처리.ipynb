```python
import pandas as pd

df = pd.read_csv('merged_reviews_with_answers1.csv')
df
```

# 버핏 -> 오버핏

```python
import ast
# keyword 열들이 이미 리스트일 경우 그대로 사용하고, 문자열일 경우 ast.literal_eval()로 리스트로 변환
for i in range(1, 5):
    keyword_col = f'keyword_{i}'
    df[f'{keyword_col}'] = df[keyword_col].apply(lambda x: [k.strip().lower() for k in x] if isinstance(x, list) else [k.strip().lower() for k in ast.literal_eval(x)] if pd.notna(x) else [])
```

```python
df.to_csv('merged_reviews_with_answers1.csv', index=False, encoding='utf-8-sig')
```

```python
df['common_keywords_3_plus'] = df['common_keywords_3_plus'].apply(lambda x: ', '.join(['오버핏' if keyword.strip() == '버핏' else keyword for keyword in str(x).split(',')]) if pd.notnull(x) else x)
df
```

```python
df['Answer'] = df['Answer'].str.lower()
```

```python
for i in range(1, 5):  # keyword_1부터 keyword_4까지 반복
    df[f'keyword_{i}'] = df[f'keyword_{i}'].apply(
        lambda x: [ '오버핏' if keyword.strip() == '버핏' else keyword for keyword in x ] if isinstance(x, list)
        else [ '오버핏' if keyword.strip() == '버핏' else keyword for keyword in str(x).split(',')] if pd.notnull(x) else x
    )
    
df.to_csv('merged_reviews_with_answers1.csv', index=False, encoding='utf-8-sig')
```

# 띄어쓰기 삭제 

```python
for i in range(1, 5):  # keyword_1부터 keyword_4까지 반복
    df[f'keyword_{i}'] = df[f'keyword_{i}'].apply(lambda x: [keyword.replace(' ', '') for keyword in x] if isinstance(x, list) else x)
df
```

```python
df.to_csv('merged_reviews_with_answers1.csv', index=False, encoding='utf-8-sig')
```

```python
# 발표자료 
import os
import pandas as pd

# 폴더 경로 설정
folder_path = 'keyword_merged_last'

# 폴더 내의 모든 CSV 파일을 읽어서 하나의 데이터프레임으로 합치기
all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.csv')]

# 모든 파일을 데이터프레임으로 읽고 합치기
df_list = [pd.read_csv(file) for file in all_files]
merged_df = pd.concat(df_list, ignore_index=True)
merged_df
```

```python
merged_df['common_keywords_3_plus'] = merged_df['common_keywords_3_plus'].apply(lambda x: ', '.join(['오버핏' if keyword.strip() == '버핏' else keyword for keyword in str(x).split(',')]) if pd.notnull(x) else x)
merged_df['common_keywords_3_plus'] = merged_df['common_keywords_3_plus'].apply(lambda x: ', '.join(['정사이즈' if keyword.strip() == '온 버핏' else keyword for keyword in str(x).split(',')]) if pd.notnull(x) else x)
merged_df['common_keywords_3_plus'] = merged_df['common_keywords_3_plus'].apply(lambda x: ', '.join(['정사이즈' if keyword.strip() == '정 사이즈' else keyword for keyword in str(x).split(',')]) if pd.notnull(x) else x)
merged_df['common_keywords_3_plus'] = merged_df['common_keywords_3_plus'].apply(lambda x: [keyword.replace(' ', '') for keyword in x] if isinstance(x, list) else x)
```

```python
merged_df = merged_df.drop(columns=['common_keywords_all'])
```

```python
# 합친 데이터프레임을 새로운 CSV 파일로 저장
merged_df.to_csv('merged_keywords.csv', index=False)

print("모든 CSV 파일이 성공적으로 하나로 합쳐졌습니다.")
```
