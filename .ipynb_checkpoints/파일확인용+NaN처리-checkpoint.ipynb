```python
'상의_결과.csv', '바지_결과.csv', '원피스_결과.csv', '아우터_결과.csv',
이렇게 파일이 4개 있고 각 파일에는
title에 구체적인 옷 이름, common_keywords_3_plus에 그 옷의 리뷰 요약한 키워드가 도출되어 있어. 이걸 보여주고 싶어.

그리고 나서 각 구체적인 옷 이름의 파일이 카테고리별로 10개씩 존재해
예를들면 '460G 컷 헤비 피그먼트 티셔츠-차콜-_keywords.csv'
여기에 원문 Review와 keywords 가 존재해서
리뷰도 출력하고 각 리뷰에 대한 keyword를 출력하게 하고 싶어.

그 다음 common_keywords_3_plus 에서 출력된 해당 옷에 대한 요약한 키워드를 기준으로 Review를 필터링 해서 보여주고 싶어.
```

```python
import pandas as pd
import ast


df = pd.read_csv('keyword_merged_last/[2PACK] 쿨 코튼 티셔츠 블랙+화이트_keywords.csv')

for i in range(1, 5):
    keyword_col = f'keyword_{i}'
    df[f'{keyword_col}'] = df[keyword_col].apply(lambda x: [k.strip().lower() for k in x] if isinstance(x, list) else [k.strip().lower() for k in ast.literal_eval(x)] if pd.notna(x) else [])

df
```

```python
import os

# 폴더 경로 설정
folder_path = 'saperate_reviews_keywords'

# 폴더 안의 파일들을 순회
for filename in os.listdir(folder_path):
    # 파일명에서 'reviews_'와 '_reviews' 문자열을 제거
    new_filename = filename.replace('reviews_', '').replace('_reviews', '')
    
    # 파일명이 변경된 경우에만 처리
    if new_filename != filename:
        # 기존 파일 경로와 새로운 파일 경로 생성
        old_file_path = os.path.join(folder_path, filename)
        new_file_path = os.path.join(folder_path, new_filename)
        
        # 파일명 변경
        os.rename(old_file_path, new_file_path)
        print(f"Renamed: {filename} -> {new_filename}")

print("All files renamed successfully!")
```

## NaN 없애기 

### 검증코드 

```python
import os
import pandas as pd

# 폴더 경로 설정
folder_path = 'saperate_reviews_tags'

# 폴더 내 '_keywords.csv'로 끝나는 파일들을 스캔하여 처리
for filename in os.listdir(folder_path):
    if filename.endswith('_1.csv'):
        file_path = os.path.join(folder_path, filename)
        
        # CSV 파일 불러오기
        df = pd.read_csv(file_path)
        
        # 'Review' 컬럼에 NaN 값이 있는지 검사
        if 'Review' in df.columns:
            nan_exists = df['Review'].isna().any()
            if nan_exists:
                print(f"NaN values found in 'Review' column of file: {filename}")
            else:
                print(f"No NaN values in 'Review' column of file: {filename}")
        else:
            print(f"'Review' column not found in file: {filename}")
```

### 삭제코드 

```python
import os
import pandas as pd

# 폴더 경로 설정
folder_path = 'saperate_reviews_tags'

# 폴더 내 '_keywords.csv'로 끝나는 파일들을 스캔하여 처리
for filename in os.listdir(folder_path):
    if filename.endswith('_1.csv'):
        file_path = os.path.join(folder_path, filename)
        
        # CSV 파일 불러오기
        df = pd.read_csv(file_path)
        
        # 'Review' 컬럼에 NaN 값이 있는지 확인 후 삭제
        if 'Review' in df.columns:
            if df['Review'].isna().any():
                # NaN 값이 있는 행 삭제
                df_cleaned = df.dropna(subset=['Review'])
                
                # NaN 값을 삭제한 데이터로 파일 덮어쓰기
                df_cleaned.to_csv(file_path, index=False)
```

## 코드에서 문서찾기
Answer200.txt

```python
import os
import json

def search_for_text_in_ipynb(directory, search_text):
    found_files = []

    # 탐색 시작
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.ipynb'):
                ipynb_path = os.path.join(root, file)
                print(f"Checking file: {ipynb_path}")  # 파일 확인 메시지 추가

                try:
                    with open(ipynb_path, 'r', encoding='utf-8') as f:
                        notebook_content = json.load(f)

                        # 노트북 파일 내 셀들을 확인
                        for cell in notebook_content.get('cells', []):
                            if cell.get('cell_type') == 'code':  # 코드 셀만 확인
                                source = ''.join(cell.get('source', []))
                                if search_text in source:
                                    print(f"Found '{search_text}' in {ipynb_path}")  # 찾으면 메시지 출력
                                    found_files.append(ipynb_path)
                                    break  # 해당 파일에서 찾았으므로 다음 파일로 이동
                except Exception as e:
                    print(f"Error reading {ipynb_path}: {e}")

    return found_files


# 사용 예시: 특정 폴더 안의 모든 ipynb 파일에서 'Answer200.txt' 검색
directory_to_search = '/your/folder/path'  # 탐색할 폴더 경로 설정
search_text = 'Answer200.txt'
found_files = search_for_text_in_ipynb(directory_to_search, search_text)

# 결과 출력
if found_files:
    print("다음 파일에서 'Answer200.txt'가 발견되었습니다:")
    for file in found_files:
        print(file)
else:
    print("'Answer200.txt'가 포함된 파일을 찾을 수 없습니다.")
```
