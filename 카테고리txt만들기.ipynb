```python
import pandas as pd
import ast
from collections import Counter

# CSV 파일 불러오기
df = pd.read_csv(file_path + '[3천원 결제혜택]링클 체크 박시 오버핏 롤업 하프 셔츠 다크 네이비_keywords.csv')

# terms_list.txt 파일에서 단어들을 리스트로 불러오기
with open('term_list.txt', 'r', encoding='utf-8') as file:
    terms_list = [line.strip().lower() for line in file.readlines()]

# common_keywords_3_plus 열에서 terms_list에 없는 단어 찾기
def find_terms_not_in_list(row, terms_list):
    if pd.notna(row):  # NaN 방지
        keywords = [term.strip().lower() for term in row.split(',')]
        # terms_list에 없는 키워드 필터링
        not_in_terms = [keyword for keyword in keywords if keyword not in terms_list]
        return ', '.join(not_in_terms) if not_in_terms else None
    return None

# 새로운 열로 terms_list에 없는 단어들 출력
df['not_in_terms'] = df['common_keywords_3_plus'].apply(lambda x: find_terms_not_in_list(x, terms_list))

# 결과 확인 (terms_list에 없는 단어들만 있는 열 출력)
# not_in_terms 열에서 NaN이 아닌 값만 선택 후 리스트로 변환
not_in_terms_list = df['not_in_terms'].dropna().tolist()

# 리스트에서 각 셀의 단어들을 모두 분리한 후 set으로 중복 제거
all_terms_not_in_list = ', '.join(not_in_terms_list).split(', ')

# 각 단어의 빈도수를 계산
term_counts = Counter(all_terms_not_in_list)

# 각 단어와 그 빈도수를 출력 (등장 횟수에 따라 정렬)
sorted_term_counts = sorted(term_counts.items(), key=lambda x: x[1], reverse=True)

# 각 단어의 빈도수 출력
print("terms_list에 없는 각 단어의 빈도수 (내림차순 정렬):")
for term, count in sorted_term_counts:
    print(f"{term}: {count}번")

# 총 단어 개수 출력
total_count = len(all_terms_not_in_list)
print(f"\n총 단어 개수: {total_count}개")

# 1번 등장한 단어 개수 출력
single_occurrence_count = sum(1 for count in term_counts.values() if count == 1)
print(f"1번 등장한 단어의 개수: {single_occurrence_count}개")
```
