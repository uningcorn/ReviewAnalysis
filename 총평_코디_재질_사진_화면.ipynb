````python
import pandas as pd
from konlpy.tag import Okt

# 파일 불러오기
file_path = 'keyword_merged_last/'
df = pd.read_csv(file_path + '데미지 워시드 데님 팬츠-미디엄 블루(Cool Air)_keywords.csv')
df = df.drop(columns=['common_keywords_all'])

# Okt 형태소 분석기
okt = Okt()

# 필터링할 키워드 목록 정의
filter_keywords = ['기장', '총기장', '기장감', '길이감', '길이', '길다', '짧다']

# 길이감 관련 키워드 사전 정의
short_keywords = ['짧다', '짧아요', '미니', '애매하다', '짧은']
long_keywords = ['길다', '길어요', '롱']

# 1. OKT 형태소 분석 및 필터링된 키워드 추출 함수
def extract_keywords(review):
    if isinstance(review, float):  # 리뷰가 NaN이거나 float일 경우 처리
        return ''
    tokens = okt.morphs(review)  # OKT 형태소 분석
    filtered_tokens = [token for token in tokens if token in filter_keywords]  # 필터링된 키워드만 추출
    return ' '.join(filtered_tokens)

# 2. 길이감 관련 키워드 분류 함수
def classify_length(keywords):
    if any(word in keywords for word in short_keywords):
        return '짧아요'
    elif any(word in keywords for word in long_keywords):
        return '길어요'
    else:
        return '적당해요'

# 'Processed_Review' 열에 대해 OKT 형태소 분석 및 필터링
df['추출된_키워드'] = df['Processed_Review'].apply(extract_keywords)

# 필터링된 키워드를 길이감에 따라 분류
df['길이감'] = df['추출된_키워드'].apply(classify_length)

# 길이감 분류에 따른 퍼센티지 계산
length_distribution = df['길이감'].value_counts(normalize=True) * 100

length_distribution
````

````python
# 형태소 분석을 수행하고, 동사, 형용사, 부사, 명사만 추출하는 함수
def analyze_morphs(review):
    if isinstance(review, float):  # NaN이나 float 값을 무시
        return ''  # 빈 문자열을 반환
    
    # 품사 태깅 (POS tagging)
    pos_tags = okt.pos(review, stem=True)
    
    # 원하는 품사만 필터링 (명사, 동사, 형용사, 부사)
    filtered_tokens = [word for word, tag in pos_tags if tag in ['Noun', 'Verb', 'Adjective', 'Adverb']]
    
    # 필터링된 단어들을 하나의 문자열로 결합해서 반환
    return ' '.join(filtered_tokens)

# 'Processed_Review' 열에 대해 형태소 분석 적용 후 새로운 열에 저장
df['Analyzed_Review'] = df['Processed_Review'].apply(analyze_morphs)
df
````

````python
def process_delivery_data(file_path):
    # CSV 파일 읽기
    df = pd.read_csv(file_path)
    
    # '버핏' -> '오버핏' 변경
    df['common_keywords_3_plus'] = df['common_keywords_3_plus'].apply(
        lambda x: ', '.join(['오버핏' if keyword.strip() == '버핏' else keyword for keyword in str(x).split(',')]) if pd.notnull(x) else x
    )
    
    # 배송 관련 키워드 리스트
    shipping_keywords = ["배송", "빠르다", "느리다", "예약배송", "깔끔하다", "예약", "오다", "하루", "이틀", "기다리다", "빨리", "일주일", "포장", "출고"]
    
    # '배송' 여부 분류
    df['delivery'] = df['common_keywords_3_plus'].apply(
        lambda x: '배송' if isinstance(x, str) and any(keyword in x for keyword in shipping_keywords) else '-'
    )
    # Okt 형태소 분석기 초기화
    okt = Okt()
    
    # 'Processed_Review_Okt' 열을 만들고, 'delivery'가 '배송'인 행만 기본형으로 변환
    df['Processed_Review_Okt'] = df['Processed_Review']
    df.loc[df['delivery'] == '배송', 'Processed_Review_Okt'] = df.loc[df['delivery'] == '배송', 'Processed_Review'].apply(
        lambda x: ' '.join([word[0] for word in okt.pos(x, stem=True)]) if isinstance(x, str) else x
    )
    
    # 'Processed_Review_Okt'에서 쉼표로 연결
    df['Processed_Review_Okt'] = df['Processed_Review_Okt'].apply(
        lambda x: ', '.join(x.split()) if isinstance(x, str) else x
    )
    
    # '배송' 관련 행만 필터링
    delivery_filtered = df[df['delivery'] == '배송'][['Processed_Review', 'common_keywords_3_plus', 'delivery', 'Processed_Review_Okt']]
    
    # 빠르다/느리다의 개수와 비율 계산
    speed_nums = delivery_filtered['speed'].value_counts()
    total_nums = len(delivery_filtered)
    speed_ratios = (speed_nums / total_nums) * 100
    speed_ratios_format = speed_ratios.apply(lambda x: f"{x:.2f}%")
    
    # '빠르다'와 '느리다' 비율만 필터링
    filter_speed_ratio = speed_ratios_format[speed_ratios_format.index.isin(['빠르다', '느리다'])]
    
    # 결과 출력
    print(f"배송 비율 (배송 관련 행 수 {total_nums}개 대비):")
    print(filter_speed_ratio)
    
    # 최종 필터된 DataFrame 반환
    return delivery_filtered

# 함수 실행 예시
file_path = './해시태그병합/DOODLE HEART HALF T WHITE GREYISH BLUE_keywords.csv'
result_df = process_delivery_data(file_path)
````

````python
# 정규표현식 활용
import re

# 찾고자 하는 단어 리스트
target_words = ['기장', '총기장', '기장김', '길이', '길다', '짧다', '미니', '롱', '길이감', '발목']
 #['화면', '이미지', '사진', '실물', '모델']  # 원하는 단어들을 여기에 추가

# 정규 표현식을 생성 ('코디|같이|포인트|매칭|매치|어울리다|활용' 형태로 만듦)
pattern = '|'.join(target_words)

# 정규 표현식을 사용하여 단어가 포함된 리뷰를 찾는 함수 정의
def find_keywords_regex(review):
    if pd.isna(review):  # NaN 값 처리
        return ''
    match = re.search(pattern, review)
    if match:
        return match.group(0)
    return ''

# 'Review' 열에서 해당 단어들이 포함된 리뷰 찾기
df['keyword_found'] = df['Review'].apply(find_keywords_regex)
````

````python
import re
import pandas as pd
from konlpy.tag import Okt

# 형태소 분석기 초기화
okt = Okt()

# 찾고자 하는 단어 리스트
target_words = ['화면', '이미지', '사진', '실물', '모델']  # 원하는 단어들을 여기에 추가

# 정규 표현식 패턴: 특정 단어 뒤에 다른 단어들이 있고 나서, 동사나 형용사를 찾음
# 단어 뒤에 여러 단어(최대 10자)를 포함한 뒤의 단어를 찾음
pattern = r'(' + '|'.join(target_words) + r')(?:\s+\S+){0,5}\s+(\S+)'

file_path = 'keyword_merged_last/'
df = pd.read_csv(file_path+'데미지 워시드 데님 팬츠-미디엄 블루(Cool Air)_keywords.csv')
df = df.drop(columns=['common_keywords_all'])

# 단어가 동사 또는 형용사인지 확인하는 함수
def is_verb_or_adjective(word):
    morphs = okt.pos(word)
    print(f"Analyzing: {word} -> {morphs}")  # 디버깅을 위해 형태소 분석 결과 출력
    for w, pos in morphs:
        if pos in ['Verb', 'Adjective']:
            return True
    return False

# 정규 표현식을 사용하여 특정 단어 뒤의 단어들 중 동사/형용사 찾기
def find_keywords_and_verbs_or_adjectives(review):
    if pd.isna(review):  # NaN 값 처리
        return ''
    matches = re.findall(pattern, review)
    print(f"Matches found in review: {matches}")  # 매칭 결과 출력
    verbs_or_adjectives = []
    for match in matches:
        word = match[1]  # 두 번째 그룹(단어)
        if is_verb_or_adjective(word):  # 해당 단어가 동사 또는 형용사인지 확인
            verbs_or_adjectives.append(word)
    return verbs_or_adjectives

# 'Review' 열에서 해당 단어 뒤에 나오는 동사/형용사 찾기
df['verbs_or_adjectives_found'] = df['Review'].apply(find_keywords_and_verbs_or_adjectives)

# 동사 또는 형용사가 포함된 리뷰만 필터링
df_filtered = df[df['verbs_or_adjectives_found'].apply(lambda x: len(x) > 0)]

# 결과 출력
df_filtered[['Review', 'verbs_or_adjectives_found']]
````
