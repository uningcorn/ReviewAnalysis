### 정확도 한꺼번에 확인하기

#### **정확도**
$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

- \(TP\): True Positives (참 양성)
- \(TN\): True Negatives (참 음성)
- \(FP\): False Positives (거짓 양성)
- \(FN\): False Negatives (거짓 음성)

```python
import pandas as pd
import ast

# CSV 파일 불러오기
df = pd.read_csv('merged_reviews_with_answers1.csv')

# Answer가 NaN이면 빈 문자열로 변환한 후 쉼표로 나누어 리스트로 변환
df['Answer_list'] = df['Answer'].apply(lambda x: [a.strip().lower() for a in str(x).split(',')] if pd.notna(x) else [])

# keyword 열들이 이미 리스트일 경우 그대로 사용하고, 문자열일 경우 ast.literal_eval()로 리스트로 변환
for i in range(1, 5):
    keyword_col = f'keyword_{i}'
    df[f'{keyword_col}_list'] = df[keyword_col].apply(lambda x: [k.strip().lower() for k in x] if isinstance(x, list) else [k.strip().lower() for k in ast.literal_eval(x)] if pd.notna(x) else [])

# 각 keyword_list와 Answer_list의 일치하는 항목 개수 및 정확도 계산
for i in range(1, 5):
    keyword_col_list = f'keyword_{i}_list'
    df[f'matches_{i}'] = df.apply(lambda row: len(set(row[keyword_col_list]) & set(row['Answer_list'])), axis=1)
    
    # 정확도는 교집합 / 합집합 으로 계산
    df[f'accuracy_{i}'] = df.apply(lambda row: len(set(row[keyword_col_list]) & set(row['Answer_list'])) / len(set(row[keyword_col_list]) | set(row['Answer_list'])) if len(set(row[keyword_col_list]) | set(row['Answer_list'])) > 0 else 0, axis=1)
    df[f'accuracy_{i}'] = df[f'accuracy_{i}'].round(4)  # 소수점 둘째 자리까지 반올림

# common_keywords_3_plus 열에 대한 처리 추가
df['3pluslist'] = df['common_keywords_3_plus'].apply(lambda x: [a.strip().lower() for a in str(x).split(',')] if pd.notna(x) else [])

# common_keywords_3_plus도 리스트로 변환하여 처리
df['3pluslist'] = df['3pluslist'].apply(lambda x: [k.strip().lower() for k in x] if isinstance(x, list) else [k.strip().lower() for k in ast.literal_eval(x)] if pd.notna(x) else [])

# common_keywords_3_plus와 Answer_list의 일치 항목 계산 및 정확도
df['matches_3plus'] = df.apply(lambda row: len(set(row['3pluslist']) & set(row['Answer_list'])), axis=1)

# 정확도는 교집합 / 합집합 으로 계산
df['accuracy_3plus'] = df.apply(lambda row: len(set(row['3pluslist']) & set(row['Answer_list'])) / len(set(row['3pluslist']) | set(row['Answer_list'])) if len(set(row['3pluslist']) | set(row['Answer_list'])) > 0 else 0, axis=1)
df['accuracy_3plus'] = df['accuracy_3plus'].round(4)

# 각 keyword별 총 정확도 계산 및 출력
for i in range(1, 5):
    total_accuracy = df[f'accuracy_{i}'].mean() * 100
    print(f'keyword_{i} 총 정확도: {total_accuracy:.4f}%')

# common_keywords_3_plus 총 정확도 계산 및 출력
total_accuracy_3plus = df['accuracy_3plus'].mean() * 100
print(f'common_keywords_3_plus 총 정확도: {total_accuracy_3plus:.4f}%')

# 원한다면 결과를 CSV로 저장 가능
# df.to_csv('accuracy_results.csv', index=False)
```

#### **f1score**

$$
\text{F1 Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
$$

#### **자카드계수**

자카드 계수(Jaccard coefficient)는 두 집합 간의 유사도를 측정하는 메트릭

$$
\text{Jaccard coefficient} = \frac{|A \cap B|}{|A \cup B|}
$$

```python
import pandas as pd
import ast

# CSV 파일 불러오기
df = pd.read_csv('merged_reviews_with_answers1.csv')

# keyword 열들이 이미 리스트일 경우 그대로 사용하고, 문자열일 경우 ast.literal_eval()로 리스트로 변환
for i in range(1, 5):
    keyword_col = f'keyword_{i}'
    df[f'{keyword_col}_list'] = df[keyword_col].apply(lambda x: [k.strip().lower() for k in x] if isinstance(x, list) else [k.strip().lower() for k in ast.literal_eval(x)] if pd.notna(x) else [])

# 자카드 계수를 계산하는 함수
def jaccard_coefficient(set1, set2):
    intersection = len(set(set1) & set(set2))  # 교집합 크기
    union = len(set(set1) | set(set2))  # 합집합 크기
    return intersection / union if union > 0 else 0

# 각 모델끼리의 자카드 계수를 계산하여 간단하게 출력
for i in range(1, 5):
    for j in range(i+1, 5):
        model_i_col = f'keyword_{i}_list'
        model_j_col = f'keyword_{j}_list'
        
        # 각 row별로 두 모델 간의 자카드 계수 계산
        df[f'jaccard_{i}_{j}'] = df.apply(lambda row: jaccard_coefficient(row[model_i_col], row[model_j_col]), axis=1)

        # 자카드 계수 평균 계산 및 간단하게 출력
        total_jaccard = df[f'jaccard_{i}_{j}'].mean() * 100
        print(f'keyword_{i} & keyword_{j} : {total_jaccard:.4f}%')

# 원한다면 결과를 CSV로 저장 가능
# df.to_csv('model_comparison_jaccard.csv', index=False)
```

```python
import pandas as pd
import ast

# CSV 파일 불러오기
df = pd.read_csv('merged_reviews_with_answers1.csv')

# keyword 열들이 이미 리스트일 경우 그대로 사용하고, 문자열일 경우 ast.literal_eval()로 리스트로 변환
for i in range(1, 5):
    keyword_col = f'keyword_{i}'
    df[f'{keyword_col}'] = df[keyword_col].apply(lambda x: [k.strip().lower() for k in x] if isinstance(x, list) else [k.strip().lower() for k in ast.literal_eval(x)] if pd.notna(x) else [])

# 공통 키워드를 추출하는 함수 정의
def extract_common_keywords_3_plus(row):
    # 각 열에서 추출한 키워드 리스트를 모두 합치기
    all_keywords = row['keyword_1'] + row['keyword_2'] + row['keyword_3'] + row['keyword_4']
    
    # 키워드의 등장 횟수 계산
    keyword_counts = pd.Series(all_keywords).value_counts()
    
    # 3번 이상 등장한 키워드를 추출
    common_keywords_3_plus = keyword_counts[keyword_counts >= 3].index.tolist()
    
    return ', '.join(common_keywords_3_plus) if common_keywords_3_plus else None

# 새로운 열에 함수 적용
df['common_keywords_3_plus'] = df.apply(extract_common_keywords_3_plus, axis=1)

# 결과 확인
df
```

# NaN 없애기 

### 검증코드 

```python
import os
import pandas as pd

# 폴더 경로 설정
folder_path = 'saperate_reviews_tags'

# 폴더 내 '_keywords.csv'로 끝나는 파일들을 스캔하여 처리
for filename in os.listdir(folder_path):
    if filename.endswith('_1.csv'):
        file_path = os.path.join(folder_path, filename)
        
        # CSV 파일 불러오기
        df = pd.read_csv(file_path)
        
        # 'Review' 컬럼에 NaN 값이 있는지 검사
        if 'Review' in df.columns:
            nan_exists = df['Review'].isna().any()
            if nan_exists:
                print(f"NaN values found in 'Review' column of file: {filename}")
            else:
                print(f"No NaN values in 'Review' column of file: {filename}")
        else:
            print(f"'Review' column not found in file: {filename}")
```

### 삭제코드 

```python
import os
import pandas as pd

# 폴더 경로 설정
folder_path = 'saperate_reviews_tags'

# 폴더 내 '_keywords.csv'로 끝나는 파일들을 스캔하여 처리
for filename in os.listdir(folder_path):
    if filename.endswith('_1.csv'):
        file_path = os.path.join(folder_path, filename)
        
        # CSV 파일 불러오기
        df = pd.read_csv(file_path)
        
        # 'Review' 컬럼에 NaN 값이 있는지 확인 후 삭제
        if 'Review' in df.columns:
            if df['Review'].isna().any():
                # NaN 값이 있는 행 삭제
                df_cleaned = df.dropna(subset=['Review'])
                
                # NaN 값을 삭제한 데이터로 파일 덮어쓰기
                df_cleaned.to_csv(file_path, index=False)
```
